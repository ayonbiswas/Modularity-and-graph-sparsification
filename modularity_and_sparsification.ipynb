```python
def read_graph(path):
    Gn = nx.read_edgelist(path,create_using = nx.Graph(),nodetype = int)
    G = nxadapter.nx2nk(Gn)
    n = G.numberOfNodes()
    m = G.numberOfEdges()
    print("nodes: ",n,"edges: ", m)
    G.toString()
    G.indexEdges()
    cc = components.ConnectedComponents(G)
    cc.run()
    print("number of components ", cc.numberOfComponents())
#     print("map of component sizes: ", cc.getComponentSizes())
    
    return G
```

```python
def sparsify(G,jt,algt,trit):
    print("-----On original graph------")
    community.detectCommunities(G, algo=community.PLM(G, True))
    
    print("-----random edge sparsification----")
    rand_spar = sparsification.RandomNodeEdgeSparsifier()
    rand_scores = rand_spar.scores(G)
    rand_graph =rand_spar._getSparsifiedGraph(G,0.2,rand_scores)
    n = rand_graph.numberOfNodes()
    m = rand_graph.numberOfEdges()
    print("nodes: ",n,"edges: ", m)
    community.detectCommunities(rand_graph, algo=community.PLM(rand_graph, True,1))
    
    print("-----forest fire sparcifiacation------")
    forest = sparsification.ForestFireSparsifier(0.7,3)
    forest_scores = forest.scores(G)
    Forest_G = forest._getSparsifiedGraph(G,10**(-19),forest_scores)
    n = Forest_G.numberOfNodes()
    m = Forest_G.numberOfEdges()
    print("nodes: ",n,"edges: ", m)
#     print(forest_scores[:1000])
    community.detectCommunities(Forest_G, algo=community.PLM(Forest_G, True,1))

    print("-----jaccard similarity sparsification ------")
    jacc = sparsification.JaccardSimilaritySparsifier()
    jacc_scores = jacc.scores(G)
    jacc_sparse = jacc._getSparsifiedGraph(G,jt,jacc_scores)
    n = jacc_sparse.numberOfNodes()
    m = jacc_sparse.numberOfEdges()
    print("nodes: ",n,"edges: ", m)
#     print(jacc_scores)
    community.detectCommunities(jacc_sparse, algo=community.PLM(jacc_sparse, True))
    
    print("-----algrebraic distance sparsification-----")
    alg = sparsification.AlgebraicDistanceSparsifier()
    alg_scores = alg.scores(G)
    alg_G = alg._getSparsifiedGraph(G,algt,alg_scores)
    n = alg_G.numberOfNodes()
    m = alg_G.numberOfEdges()
    print("nodes: ",n,"edges: ", m)
#     print(alg_scores[:1000])
    community.detectCommunities(alg_G, algo=community.PLM(alg_G, True))
        
    print("-----triangle sparsification ------")
    tri_spar = sparsification.TriangleSparsifier()
    tri_scores = tri_spar.scores(G)
    tri_graph =tri_spar._getSparsifiedGraph(G,trit,tri_scores)
    n = tri_graph.numberOfNodes()
    m = tri_graph.numberOfEdges()
    print("nodes: ",n,"edges: ", m)
    community.detectCommunities(tri_graph, algo=community.PLM(tri_graph, True))
    

#     print("-----specral sparsification ------")
#     adj = networkit.algebraic.adjacencyMatrix(G)
#     G_ = pygsp.graphs.Graph(adj)
#     epsilon = 0.4
#     G2 = reduction.graph_sparsify(G_, epsilon)
#     A = G2.W.todense()
#     G_nx =nx.from_numpy_matrix(A)
#     G_nk = nxadapter.nx2nk(G_nx)
#     community.detectCommunities(G_nk, algo=community.PLM(G_nk, True))
```
